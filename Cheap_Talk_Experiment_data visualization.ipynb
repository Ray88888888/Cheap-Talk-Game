````python
dt2s = dt2[dt2.player_id == 1]

summary.loc['dt2', 'N_good'] = dt2s[(dt2s['value1'] >= 60)].shape[0] \
                               + dt2s[(dt2s['value2'] >= 60)].shape[0]

summary.loc['dt2', 'N_bad'] = dt2s[(dt2s['value1'] < 60)].shape[0] \
                               + dt2s[(dt2s['value2'] < 60)].shape[0]

summary.loc['dt2', 'rec_good'] = (dt2s[(dt2s.value1 >= 60) & (dt2s.message1 == 1)].shape[0] +
              dt2s[(dt2s.value2 >= 60) & (dt2s.message2 == 1)].shape[0]) / \
             (dt2s[(dt2s.value1 >= 60)].shape[0] + dt2s[(dt2s.value2 >= 60)].shape[0])

summary.loc['dt2', 'rec_bad'] = (dt2s[(dt2s.value1 < 60) & (dt2s.message1 == 1)].shape[0] +
              dt2s[(dt2s.value2 < 60) & (dt2s.message2 == 1)].shape[0]) / \
             (dt2s[(dt2s.value1 < 60)].shape[0] + dt2s[(dt2s.value2 < 60)].shape[0])


summary.loc['dt2', 'take_rec'] = (dt2s[(dt2s['take1'] == 1) & (dt2s.message1 == 1)].shape[0] +
              dt2s[(dt2s['take2'] == 1) & (dt2s.message2 == 1)].shape[0]) / \
             (dt2s[dt2s.message1 == 1].shape[0] + dt2s[dt2s.message2 == 1].shape[0])

summary.loc['dt2', 'take_norec'] = (dt2s[(dt2s['take1'] == 1) & (dt2s.message1 == 2)].shape[0] +
                dt2s[(dt2s['take2'] == 1) & (dt2s.message2 == 2)].shape[0]) / \
                (dt2s[dt2s.message1 == 2].shape[0] + dt2s[dt2s.message2 == 2].shape[0]) \
    if (dt2s[dt2s.message1 == 2].shape[0] + dt2s[dt2s.message2 == 2].shape[0]) > 0 else 999

summary.loc['dt2', 'take_good'] = (dt2s[(dt2s['take1'] == 1) & (dt2s.value1 >= 60)].shape[0] +
                 dt2s[(dt2s['take2'] == 1) & (dt2s.value2 >= 60)].shape[0]) / \
                (dt2s[dt2s.value1 >= 60].shape[0] + dt2s[dt2s.value2 >= 60].shape[0])

summary.loc['dt2', 'take_bad'] = (dt2s[(dt2s['take1'] == 1) & (dt2s.value1 < 60)].shape[0] +
                 dt2s[(dt2s['take2'] == 1) & (dt2s.value2 < 60)].shape[0]) / \
                (dt2s[dt2s.value1 < 60].shape[0] + dt2s[dt2s.value2 < 60].shape[0])
````

````python
dt3s = dt3[dt3.player_id == 1]

summary.loc['dt3', 'N_good'] = dt3s[(dt3s['value1'] >= 60)].shape[0] \
                               + dt3s[(dt3s['value2'] >= 60)].shape[0] + dt3s[(dt3s['value3'] >= 60)].shape[0]

summary.loc['dt3', 'N_bad'] = dt3s[(dt3s['value1'] < 60)].shape[0] \
                               + dt3s[(dt3s['value2'] < 60)].shape[0] + dt3s[(dt3s['value3'] < 60)].shape[0]

summary.loc['dt3', 'rec_good'] = (dt3s[(dt3s.value1 >= 60) & (dt3s.message1 == 1)].shape[0] +
                                  dt3s[(dt3s.value2 >= 60) & (dt3s.message2 == 1)].shape[0] +
                                  dt3s[(dt3s.value3 >= 60) & (dt3s.message3 == 1)].shape[0]) / \
                                 (dt3s[(dt3s.value1 >= 60)].shape[0] + dt3s[(dt3s.value2 >= 60)].shape[0] +
                                  dt3s[(dt3s.value3 >= 60)].shape[0])

summary.loc['dt3', 'rec_bad'] = (dt3s[(dt3s.value1 < 60) & (dt3s.message1 == 1)].shape[0] +
                                 dt3s[(dt3s.value2 < 60) & (dt3s.message2 == 1)].shape[0] +
                                 dt3s[(dt3s.value3 < 60) & (dt3s.message3 == 1)].shape[0]) / \
                                (dt3s[(dt3s.value1 < 60)].shape[0] + dt3s[(dt3s.value2 < 60)].shape[0] +
                                 dt3s[(dt3s.value3 < 60)].shape[0])


summary.loc['dt3', 'take_rec'] = (dt3s[(dt3s['take1'] == 1) & (dt3s.message1 == 1)].shape[0] +
                                  dt3s[(dt3s['take2'] == 1) & (dt3s.message2 == 1)].shape[0] +
                                  dt3s[(dt3s['take3'] == 1) & (dt3s.message3 == 1)].shape[0]) / \
                                 (dt3s[dt3s.message1 == 1].shape[0] + dt3s[dt3s.message2 == 1].shape[0] +
                                  dt3s[dt3s.message3 == 1].shape[0])

summary.loc['dt3', 'take_norec'] = (dt3s[(dt3s['take1'] == 1) & (dt3s.message1 == 2)].shape[0] +
                                    dt3s[(dt3s['take2'] == 1) & (dt3s.message2 == 2)].shape[0] +
                                    dt3s[(dt3s['take3'] == 1) & (dt3s.message3 == 2)].shape[0]) / \
                                   (dt3s[dt3s.message1 == 2].shape[0] + dt3s[dt3s.message2 == 2].shape[0] +
                                    dt3s[dt3s.message3 == 2].shape[0]) \
    if (dt3s[dt3s.message1 == 2].shape[0] + dt3s[dt3s.message2 == 2].shape[0] +dt3s[dt3s.message3 == 2].shape[0]) > 0 else 999

summary.loc['dt3', 'take_good'] = (dt3s[(dt3s['take1'] == 1) & (dt3s.value1 >= 60)].shape[0] +
                                   dt3s[(dt3s['take2'] == 1) & (dt3s.value2 >= 60)].shape[0] +
                                   dt3s[(dt3s['take3'] == 1) & (dt3s.value3 >= 60)].shape[0]) / \
                                  (dt3s[dt3s.value1 >= 60].shape[0] + dt3s[dt3s.value2 >= 60].shape[0] +
                                   dt3s[dt3s.value3 >= 60].shape[0])

summary.loc['dt3', 'take_bad'] = (dt3s[(dt3s['take1'] == 1) & (dt3s.value1 < 60)].shape[0] +
                                  dt3s[(dt3s['take2'] == 1) & (dt3s.value2 < 60)].shape[0] +
                                  dt3s[(dt3s['take2'] == 1) & (dt3s.value2 < 60)].shape[0]) / \
                                 (dt3s[dt3s.value1 < 60].shape[0] + dt3s[dt3s.value2 < 60].shape[0] +
                                  dt3s[dt3s.value3 < 60].shape[0])
````

##Variable Intro
---
```markdown
* 'N_good', number of good items
* 'N_bad', number of bad items
* 'rec_good', P(A recommends|item is good)
* 'rec_bad', P(A recommends|item is bad)
* 'take_rec', P(B acquires | A recommends )
* 'take_norec', P(B acquires | A not recommends )
* 'take_good', P(B acquires | item good )
* 'take_bad',  P(B acquires | item bad )
```
---
For the 3 graphs, we are interested in:
(1) //lying ratio//  'rec_bad', P(A recommends|item is bad)
(2) //trust ratio//  'take_rec', P(B acquires | A recommends )
(3) //accuracy of B//  ( a weighted average of P(B acquires | item good ) and P(B not acquires | item bad ))

````python
Variable = ['round','N_itemisbad','N_rec_bad','rec_bad','N_rec','N_take_rec','take_rec',
            'N_itemisgood','N_take_good','take_good','N_notake_bad', 'notake_bad']
Index = ['dt1']*10+['dt2']*10+['dt3']*10
df = pd.DataFrame(index=Index, columns=Variable)
df.head()
````

##Accuracy of B

Accuracy of B = P(B acquires | item good )*(item good/total item) + P(B not acquires|item bad)*(item bad/total item) 

````python
df.insert(12, "B_accuracy", [0]*30, True)
````

````python
df['N_itemisgood']=[0]*30
df['N_take_good']=[0]*30
df['N_notake_bad']=[0]*30
````

````python
df.head()
````

````python
dt1s.head()
````

````python
#caculate Accuracy of B in Treatment One (dt1)
index=dt1s.index
for k in index:
  #B not acquires|item bad
  if dt1s['value'][k] < 60:
    roundnumber=dt1s['round'][k]-1
    if dt1s['take'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1

  #B acquires|item good
  if dt1s['value'][k] >= 60:
    roundnumber=dt1s['round'][k]-1
    df['N_itemisgood'][roundnumber]+=1
    if dt1s['take'][k]== 1:
      df['N_take_good'][roundnumber]+=1

df['notake_bad'] = df['N_notake_bad']/df['N_itemisbad']
df['take_good'] = df['N_take_good']/df['N_itemisgood']
df.head()
````

````python
dt2s.head()
````

````python
#caculate Accuracy of B in Treatment Two (dt2)
index=dt2s.index
for k in index:
  #B not acquires|item bad
  if dt2s['value1'][k] < 60:
    roundnumber=dt2s['round'][k]+9
    if dt2s['take1'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1

  if dt2s['value2'][k] < 60:
    roundnumber=dt2s['round'][k]+9
    if dt2s['take2'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1


  #B acquires|item good
  if dt2s['value1'][k] >= 60:
    roundnumber=dt2s['round'][k]+9
    df['N_itemisgood'][roundnumber]+=1
    if dt2s['take1'][k]== 1:
      df['N_take_good'][roundnumber]+=1

  if dt2s['value2'][k] >= 60:
    roundnumber=dt2s['round'][k]+9
    df['N_itemisgood'][roundnumber]+=1
    if dt2s['take2'][k]== 1:
      df['N_take_good'][roundnumber]+=1

df['notake_bad'] = df['N_notake_bad']/df['N_itemisbad']
df['take_good'] = df['N_take_good']/df['N_itemisgood']
df.head()
````

````python
#caculate Accuracy of B in Treatment Three (dt3)
index=dt3s.index
for k in index:
  #B not acquires|item bad
  if dt3s['value1'][k] < 60:
    roundnumber=dt3s['round'][k]+19
    if dt3s['take1'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1

  if dt3s['value2'][k] < 60:
    roundnumber=dt3s['round'][k]+19
    if dt3s['take2'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1
  
  if dt3s['value3'][k] < 60:
    roundnumber=dt3s['round'][k]+19
    if dt3s['take3'][k]== 2:
      df['N_notake_bad'][roundnumber]+=1


  #B acquires|item good
  if dt3s['value1'][k] >= 60:
    roundnumber=dt3s['round'][k]+19
    df['N_itemisgood'][roundnumber]+=1
    if dt3s['take1'][k]== 1:
      df['N_take_good'][roundnumber]+=1

  if dt3s['value2'][k] >= 60:
    roundnumber=dt3s['round'][k]+19
    df['N_itemisgood'][roundnumber]+=1
    if dt3s['take2'][k]== 1:
      df['N_take_good'][roundnumber]+=1
  
  if dt3s['value3'][k] >= 60:
    roundnumber=dt3s['round'][k]+19
    df['N_itemisgood'][roundnumber]+=1
    if dt3s['take3'][k]== 1:
      df['N_take_good'][roundnumber]+=1

df['notake_bad'] = df['N_notake_bad']/df['N_itemisbad']
df['take_good'] = df['N_take_good']/df['N_itemisgood']
df.head()
````

````python
#calculate the Accuracy of B
df['B_accuracy']= (df['take_good']*(df['N_itemisgood']/(df['N_itemisgood']+df['N_itemisbad']))) + (df['notake_bad']*(df['N_itemisbad']/(df['N_itemisgood']+df['N_itemisbad'])))
````
